## Regression

```python
w = 0.7 
b = 0.3

start = 0
end = 1
step = .02
X = torch.arange(start, end, step).unsqueeze(dim=1)

y = w * X + b
```

```python
X[:10], y[:10]
```

```python
# train, test split 
train_split = int(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]

len(X_train), len(y_train)
len(X_test), len(y_test)
```

```python
X_train, y_train
```

```python
import matplotlib.pyplot as plt

def plot_predictions(train_data = X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=None):
    """
    Plots train and test data, comapare predictons
    """

    plt.figure(figsize=(10,7))

    # train in blue 
    plt.scatter(train_data, train_labels, c = "b", s=4, label="Train data")

    # test data in green 
    plt.scatter(test_data, test_labels, c="g", s=4, label="Test")

    if (predictions is not None):
        plt.scatter(test_data, predictions, c="r", s = 4,label="predictions")
    
    plt.legend(prop={"size": 14})
    plt.grid()
    
```

```python
plot_predictions()
```

```python
from torch import nn

class LinearRegression(nn.Module):

    def __init__(self):
        super().__init__()
        self.weights = nn.Parameter(torch.randn(1,
                                                requires_grad=True,
                                                dtype=torch.float))
        self.bias = nn.Parameter(torch.randn(1,
                                             requires_grad=True,
                                             dtype=torch.float))
        
    # Forward pass 
    def forward(self, x: torch.Tensor) -> torch.tensor: 
        return self.weights * x + self.bias
```

```python
torch.manual_seed(seed)

model0 = LinearRegression()

list(model0.parameters())
model0.state_dict()
```

```python
with torch.inference_mode():
    y_preds = model0(X_test)

y_preds
```

```python
plot_predictions(predictions=y_preds)
```

```python
# mean absolute error for loss (MAE)
loss_fn = nn.L1Loss()

# optimimzer (stochastic gradient descent)
optimizer = torch.optim.SGD(params=model0.parameters(),
                            lr=.01)
```

```python
# Train loop 
epochs = 200

epoch_count = []
train_loss_vals = []
test_loss_vals = []

for epoch in range(epochs):

    # training mode 
    model0.train()

    # 1. forward pass
    y_pred = model0(X_train)

    # 2. calculate loss
    loss = loss_fn(y_pred, y_train)
    
    # 3. opt zero grad 
    optimizer.zero_grad()

    # 4. backprop wrt. params of model
    loss.backward()

    # 5. step optimizer 
    optimizer.step()

    # Testing 
    model0.eval()
    with torch.inference_mode():
        test_pred = model0(X_test)

        # Calculate loss 
        test_loss = loss_fn(test_pred, y_test)
    
    if (epoch % 10 == 0):
        print(f"Epoch: {epoch} | Loss: {loss} | Test loss {test_loss}")

    epoch_count.append(epoch)
    train_loss_vals.append(loss)
    test_loss_vals.append(test_loss)
    
print(model0.state_dict())
```

```python
class LinearRegressionV2(nn.Module):

    def __init__(self):
        super().__init__()
        # Use nn.Linear() for model params: has weight and bias
        self.linear_layer = nn.Linear(in_features=1,
                                      out_features=1)
        
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.linear_layer(x)
    

torch.manual_seed(seed)
model1 = LinearRegressionV2()
model1.state_dict(), model1


```

```python
def accuracy_fn(y_true, y_pred):
    correct = torch.eq(y_true, y_pred).sum().item()
    acc = (correct/len(y_pred)) * 100
    return acc
```

```python
class CircleModelV1(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(in_features=2, out_features=10)
        self.layer2 = nn.Linear(in_features=10, out_features=10)
        self.layer3 = nn.Linear(in_features=10, out_features=1)

    def forward(self, x):
        return self.layer3(self.layer2(self.layer1(x)))

model1 = CircleModelV1().to(device)
model1
```

```python
# Loss function 
loss_fn = nn.BCEWithLogitsLoss()

# Create optimizer
optimizer = torch.optim.SGD(params=model1.parameters(),
                            lr=0.1)
```

```python
# Training loop

torch.manual_seed(seed)
torch.mps.manual_seed(seed)

epochs = 1000

# Move to mps 
X_train, y_train = X_train.to(device), y_train.to(device)
X_test, y_test = X_test.to(device), y_test.to(device)

for epoch in range(epochs):

    # Train
    model1.train()

    # 1. Forward pass
    y_logits = model1(X_train).squeeze()
    y_pred = torch.round(torch.sigmoid(y_logits)) # logits->pred probabilities -> pred labels

    # 2. Loss, acc
    loss = loss_fn(y_logits, y_train)
    acc = accuracy_fn(y_true=y_train,
                      y_pred=y_pred)
    
    # 3. Zero optimizer 
    optimizer.zero_grad()

    # 4. Backprop 
    loss.backward()

    # 5. Step optimizer (GD)
    optimizer.step()

    # Testing
    model1.eval()
    with torch.inference_mode():
        # 1. Forward pass 
        test_logits = model1(X_test).squeeze()
        test_pred = torch.round(torch.sigmoid(test_logits))

        # 2. Loss, acc
        test_loss = loss_fn(test_logits,
                            y_test)
        test_acc = accuracy_fn(y_true=y_test,
                               y_pred=test_pred)
        
    if (epoch % 100 == 0):
        print(f"Epoch: {epoch} | Loss: {loss:.5f}, Acc: {acc:.2f} | Test Loss: {test_loss:.5f}, Test Acc: {test_acc:.2f}" )

```

## Replicate nonlinear activations

```python
A = torch.arange(-10,10,1, dtype=torch.float)
plt.plot(A)
```

```python
plt.plot(torch.relu(A))
```

```python
def relu(x: torch.tensor) -> torch.tensor:
    return torch.maximum(torch.tensor(0),x)

plt.plot(relu(A))
```

```python
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))
```

```python
plt.plot(sigmoid(A))
```
